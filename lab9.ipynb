```python
import nltk
from nltk import CFG
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N | Det N PP
VP -> V NP | VP PP
PP -> P NP
Det -> 'the' | 'a'
N -> 'teacher' | 'concepts' | 'students' | 'lecture'
V -> 'explained'
P -> 'to' | 'during'
""")
sentence = "the teacher explained the concepts to the students during the lecture"
tokens = list(sentence.split())
top_down_parser = nltk.RecursiveDescentParser(grammar)
bottom_up_parser = nltk.ShiftReduceParser(grammar)
print("Top-Down Parsing Trees:")
for tree in top_down_parser.parse(tokens):
    print(tree)
    tree.draw()
print("Bottom-Up Parsing Trees:")
for tree in bottom_up_parser.parse(tokens):
    print(tree)
    tree.draw()
```
