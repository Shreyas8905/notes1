```python
import random
def complete_sentence(sentence_template, options):
    if "____" not in sentence_template:
        raise ValueError("The sentence must contain a blank represented by '____'.")
    completion = random.choice(options)
    completed_sentence = sentence_template.replace("____", completion)
    return completed_sentence
sentence = "the sun rises in the ____"
options = ["east", "west", "morning sky", "horizon"]
for _ in range(5):
    print(complete_sentence(sentence, options))
```
