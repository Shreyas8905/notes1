```python
sentence = "the teacher explained the difficult concepts to the students during the lecture"
tokens = sentence.split()
verbs = ["explained", "taught", "showed", "gave", "discussed"]
prepositions = ["to", "during", "in", "on", "with", "for", "at"]
subject = None
verb = None
objects = []
prep_phrases = []
i = 0
while i < len(tokens):
    token = tokens[i]
    if not subject and token not in verbs and token not in prepositions and token != "the":
        subject = token
    if token in verbs:
        verb = token
    elif verb and token not in prepositions and token != "the":
        objects.append(token)
    if token in prepositions:
        phrase = [token]
        i += 1
        while i < len(tokens) and tokens[i] != "the" and tokens[i] not in prepositions:
            phrase.append(tokens[i])
            i += 1
        prep_phrases.append(" ".join(phrase))
        continue  

    i += 1
print("Subject:", subject)
print("Verb:", verb)
print("Objects:", objects)
print("Prepositional Phrases:", prep_phrases)
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```python
sentence = "the teacher explained the difficult concepts to the students during the lecture"
tokens = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(tokens)
grammar = r"""
  NP: {<DT>?<JJ>*<NN.*>}   
  VP: {<VB.*><NP|PP>*}     
  PP: {<IN><NP>}           
"""
cp = nltk.RegexpParser(grammar)
tree = cp.parse(pos_tags)
for subtree in tree.subtrees():
    if subtree.label() in ["NP", "VP", "PP"]:
        print(f"{subtree.label()}: {' '.join(word for word, tag in subtree.leaves())}")
```
